## 1.1 优化的目标

```python
%matplotlib inline
import numpy as np
import tensorflow as tf
from mpl_toolkits import mplot3d
import d2l
```

优化与深度学习的目标是不同的，前者关心的是最小化目标，后者关注的是在给定的有限数据量的情况下寻找合适的模型。

经验风险：训练数据集的平均损失；\
风险：整个数据群的预期损失。

```python
def f(x):
    return x*tf.cos(np.pi*x)

def g(x):
    return f(x)+0.2*tf.cos(5*np.pi*x)
```

```python
def annotate(text,xy,xytext):
    d2l.plt.gca().annotate(text,xy=xy,xytext=xytext,arrowprops=dict(arrowstyle='->'))

x = tf.range(0.5,1.5,0.01)
d2l.set_figsize((4.5,2.5))
d2l.plot(x,[f(x),g(x)],'x','risk')
annotate('min of \nempirical risk',(1.0, -1.2), (0.5, -1.1))
annotate('min of risk', (1.1, -1.05), (0.95, -0.5))
```

# 2. 凸性

凸集：如果对于任何$a,b \in \mathcal{X}$，连接$a$和$b$的线段也位于$\mathcal{X}$中，则向量空间中的一个集合$\mathcal{X}$的凸的。

詹森不等式：
给定一个凸函数$f$
$$
\sum_i \alpha_i f(x_i) \ge f(\sum_i \alpha_i x_i) \\
E_X[f(X)] \ge f(E_X[X])
$$
其中，$\alpha_i$是非负实数，因此$\sum_i \alpha_i =1$。\
凸函数的期望不小于期望的凸函数。
